```python
x = [-3, 5, 7]
y = [10, 2, 5]
```

```python
fig = plt.figure(figsize=(8,3))
plt.plot(x,y)

plt.xlim(0, 10)

plt.ylim(-3, 8)
plt.xlabel('Sales',color='blue')
plt.ylabel('product')
plt.title('Line Plot')
plt.suptitle('Sales Comparison', size=20, y=1.05);

#plt.legend()
#plt.show()
```

# seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
```

```python
tips.head()
```

## Univariate  one variable data visualization

```python
#strip plot
sns.stripplot(x='sex',y='tip', data=tips, jitter = True)
plt.title('Strip plot')
plt.xlabel('Gender')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='sex', y='tip',hue='day', data=tips)
plt.ylabel('Tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, hue = "sex",jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

```python
tips
```

```python
plt.hist(tips['total_bill'])
# ax[1].boxplot()
```

```python
plt.hist(tips['tip'])
```

```python
# subplots
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2)
sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

## Bivariate  two variables data visualization

```python
#Joint plot

sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
# Density plot
sns.jointplot(x='total_bill', y= 'tip', data=tips,kind='kde')
plt.show()
# Contours are created representing various levels of density of the data
```

```python
pd.plotting.scatter_matrix(tips)
plt.show()
```

```python
#Pair plot
sns.pairplot(tips,hue = "sex")
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```
